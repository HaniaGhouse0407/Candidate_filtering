```python

def filtering_applicants(dataset_path,start_req,python_req,ml_req,nlp_req,dl_req):

  df = pd.read_csv(dataset_path)
  filtered_df = df[
      (df["Are you available for 3 months, starting immediately, for a full-time work from home internship? "] == start_req) &
       (df['Python (out of 3)'] >= python_req) &
        (df['Machine Learning (out of 3)'] >= ml_req) &
         (df['Natural Language Processing (NLP) (out of 3)'] >= nlp_req)&
          (df['Deep Learning (out of 3)'] >= dl_req)

    ]
  filtered_df = filtered_df.drop(['Name', 'Performance_PG', 'Performance_UG', 'Performance_12', 'Performance_10'], axis=1)  # Delete specified columns
  filtered_df['ID'] = 0
  counter = 1
  for index, row in filtered_df.iterrows():
    filtered_df.at[index, 'ID'] = counter
    counter += 1

  return filtered_df

dataset_path= "/content/Applications_for_Machine_Learning_internship_edited.xlsx - Sheet1.csv"
start_req= "Yes, I am available for 3 months starting immediately for a full-time internship."
python_req= 3
ml_req= 2
nlp_req= 1
dl_req=1
filtered_applicants = filtering_applicants(dataset_path,start_req,python_req,ml_req,nlp_req,dl_req)
#print(filtered_applicants)
filtered_applicants.to_csv('filtered_dataset.csv', index=False)
df1=pd.read_csv("/content/filtered_dataset.csv")
df1
```

In this piece of code we are calculating the scores which will help us identify the best candidate .


```python
def calculate_score(row, python_req, ml_req, nlp_req, dl_req, other_skills_weight):
    score = 0

    # Technical skills
    score += row['Python (out of 3)'] * python_req
    score += row['Machine Learning (out of 3)'] * ml_req
    score += row['Natural Language Processing (NLP) (out of 3)'] * nlp_req
    score += row['Deep Learning (out of 3)'] * dl_req

    # Other relevant skills with weights
    if 'Git' in row['Other skills']:
        score += 1 * other_skills_weight
    if 'Docker' in row['Other skills']:
        score += 1 * other_skills_weight
    if 'R' in row['Other skills']:
        score += 1 * other_skills_weight
    if 'SQL' in row['Other skills']:
        score += 1 * other_skills_weight
    if 'Data Science' in row['Other skills']:
        score += 1 * other_skills_weight
    if 'C++ Programming' in row['Other skills']:
        score += 1 * other_skills_weight
    if 'Java' in row['Other skills']:
        score += 1 * other_skills_weight
    if 'Unix' in row['Other skills']:
        score += 1 * other_skills_weight
    if ' Flask' in row['Other skills']:
        score += 1 * other_skills_weight
    if 'Amazon Web Services (AWS)' in row['Other skills']:
        score += 1 * other_skills_weight
    if ' OpenCV' in row['Other skills']:
        score += 1 * other_skills_weight
    if ' Torch' or 'PyTorch' in row['Other skills']:
        score += 1 * other_skills_weight
    if ' Tensorflow' in row['Other skills']:
        score += 1 * other_skills_weight

    if ' Django' in row['Other skills']:
        score += 1 * other_skills_weight
    if ' Hadoop' in row['Other skills']:
        score += 1 * other_skills_weight
    if ' MongoDB' in row['Other skills']:
        score += 1 * other_skills_weight



    return score

def find_best_applicant(df, python_req, ml_req, nlp_req, dl_req, year_req, other_skills_weight):
    # Filter applicants based on year of graduation
    filtered_df = df[df['Current Year Of Graduation'] <= year_req]

    # Calculate scores for each applicant
    filtered_df['Score'] = filtered_df.apply(lambda row: calculate_score(row, python_req, ml_req, nlp_req, dl_req, other_skills_weight), axis=1)

    # Find the applicant with the highest score
    best_applicant = filtered_df.loc[filtered_df['Score'].idxmax()]
    print("Best Candidate:")
    print("ID:", best_applicant['ID'])
    print("Skill Levels:")
    print("Python:", best_applicant['Python (out of 3)'])
    print("Machine Learning:", best_applicant['Machine Learning (out of 3)'])
    print("Natural Language Processing (NLP):", best_applicant['Natural Language Processing (NLP) (out of 3)'])
    print("Deep Learning:", best_applicant['Deep Learning (out of 3)'])
    print("Other Skills:", best_applicant['Other skills'])
    print("Current Year Of Graduation:", best_applicant['Current Year Of Graduation'])
    print("Availibility:", best_applicant['Are you available for 3 months, starting immediately, for a full-time work from home internship? '])
    print("Degree:", best_applicant['Degree'])
    print("Stream:", best_applicant['Stream'])


    return best_applicant

# Example usage
python_req = 3
ml_req = 2
nlp_req = 1
dl_req = 1
year_req = 2024
other_skills_weight = 0.5

best_applicant = find_best_applicant(filtered_applicants, python_req, ml_req, nlp_req, dl_req, year_req, other_skills_weight)

```
